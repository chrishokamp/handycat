```python
import json
import copy
from collections import OrderedDict
```

```python
project_urls = {
    "data/word_level_qe_experiments/experiment_data/xliff/test_0.xliff",
    "data/word_level_qe_experiments/experiment_data/xliff/test_1.xliff",
    "data/word_level_qe_experiments/experiment_data/xliff/test_2.xliff",
    "data/word_level_qe_experiments/experiment_data/xliff/test_3.xliff",
    "data/word_level_qe_experiments/experiment_data/xliff/test_4.xliff"
}
```

```python
template = OrderedDict([
  ("name", ""),
  ("internalName", ""),
  ("url", ""),
  ("configuration", OrderedDict([
    ("target", OrderedDict([
      ("services", OrderedDict([
        ("useConstrainedDecoding", False),
        ("useWordLevelQE", False),
      ])),
      ("activeComponent", "wordLevelQeEditor"),
      ("defaultComponent", "wordLevelQeEditor"),
      ("components", [
        OrderedDict([
          ("directiveName", "wordLevelQeEditor"),
          ("textName", "wordLevelQeEditor"),
        ])
      ]),
    ])),
  ]))
])
```

```python
sample_files = []

for project_letter, project_url in zip(['A', 'B', 'C', 'D', 'E'], project_urls):
    project_basename = project_url.split('/')[-1].split('.')[0]
    for config in ['DEMO', 'BASIC', 'QE', 'CD', 'IPE']:
        new_project = copy.deepcopy(template)
        new_project["url"] = project_url
        new_project["name"] = '{}-Task-{}'.format(config, project_letter)
        new_project["internalName"] = '{}-{}'.format(project_basename, config)
        if config == 'DEMO':
            new_project['configuration']['target']['services']['useConstrainedDecoding'] = True
            new_project['configuration']['target']['services']['useWordLevelQE'] = True
        elif config == 'BASIC':
            new_project['configuration']['target']['services']['useConstrainedDecoding'] = False
            new_project['configuration']['target']['services']['useWordLevelQE'] = False
        elif config == 'QE':
            new_project['configuration']['target']['services']['useConstrainedDecoding'] = False
            new_project['configuration']['target']['services']['useWordLevelQE'] = True
        elif config == 'CD':
            new_project['configuration']['target']['services']['useConstrainedDecoding'] = True
            new_project['configuration']['target']['services']['useWordLevelQE'] = False
        elif config == 'IPE':
            new_project['configuration']['target']['services']['useConstrainedDecoding'] = True
            new_project['configuration']['target']['services']['useWordLevelQE'] = True
        else:
            raise ValueError('Unknown config type: {}'.format(config))
        sample_files.append(new_project)
```

```python
print(json.dumps(sample_files, indent=2))
```
