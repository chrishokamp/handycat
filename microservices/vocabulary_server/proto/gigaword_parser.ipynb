```python
from __future__ import division, print_function
import numpy as np
import codecs
import re

%load_ext autoreload
%autoreload 2
```

```python
gigaword_file = 'afp_eng_199405'

output_lines = []
with codecs.open(gigaword_file) as raw:
    for l in raw:
#         remove all tags
        l = re.sub('<.*>', '', l)
        output_lines.append(l.decode('utf8'))
```

```python
just_text_lines = []
for idx,l in enumerate(output_lines):
    # if this isn't a 'list item'
    # todo, make a list of regex conditions and use 'all'
    if not re.match('^\s{2,}', l):
        # if this isn't a blank line
        if not re.match('^$', l):
            just_text_lines.append(l.strip())
#             if the next line is blank, add a split
            if re.match('^$', output_lines[idx+1]):
                just_text_lines.append('\n')
```

```python
# what's the maximum line length?
maxlen = np.argmax([len(l) for l in just_text_lines])
maxlen
just_text_lines[maxlen]
```

```python
# append lines together, then split on newlines again
full_texts = [l.strip() for l in " ".join(just_text_lines).split('\n')]
len(full_texts)
```

```python
# prune conditions
def is_not_empty_line(line):
    if len(line) == 0:
        return False
    else:
        return True
    
# a sentence needs at least 3 tokens separated by whitespace
def is_not_too_short(line, mintoks=3):
    matches = re.findall('\s+', line)
    if len(matches) < mintoks:
        return False
    else:
        return True

# lines which contain too many numbers
def too_many_nums(line, cutoff=3):
    matches = re.findall('[0-9]+', line)
    if len(matches) > cutoff:
        return False
    else:
        return True
    
def too_many_caps_words(line, cutoff=3):
    matches = re.findall('\s+[A-Z]+\s+', line)
    if len(matches) > cutoff:
        return False
    else:
        return True
    
# sentences which start with a capitalized word tend to be something weird
def starts_with_caps_word(line, minlen=3):
#     matches = re.match('^[A-Z]{' + str(minlen) + ',}\s+', line)
    matches = re.match('^[A-Z]{' + str(minlen) + ',}', line)
    if matches:
        return False
    else:
        return True
    
# MORE PRUNE CONDITIONS HERE
    
prune_conditions = [is_not_too_short, is_not_empty_line, too_many_nums, too_many_caps_words, starts_with_caps_word]
```

```python
good_lines = [l for l in full_texts if all([cond(l) for cond in prune_conditions])] 
```
