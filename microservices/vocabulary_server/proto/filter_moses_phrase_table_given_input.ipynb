```python
from __future__ import division, print_function
import numpy as np
import codecs
import re

%load_ext autoreload
%autoreload 2
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
from lm_autocomplete import extract_phrases
```

```python
# very simple vocab filter for a moses pt
# TODO use gzip and manually encode to utf8
input_file = '/home/chris/Desktop/Dropbox/projects/malaga_lm_autocompleters_april_2015/experiment_data/malaga_experiment_segments.txt'


filter_set = set()
with codecs.open(input_file, encoding='utf8') as input_file:
    for l in input_file:
        l = l.strip()
        line_toks = wordpunct_tokenize(l)
        phrases = [tuple(p) for p in extract_phrases(line_toks, max_len=3)]
        filter_set.update(phrases)
# filter by vocab and other constraints
# if phrase doesn't begin and end with a word in the vocab, don't keep it

# load and tokenize the input filters

# get the set of all phrases in input up to size N

#
```

```python
# load the phrase table, tokenize the source part
full_phrase_table = '/home/chris/projects/angular/editor_components/microservices/lm_autocomplete/test_data/phrase_tables/wmt-phrase-table.en-es.filtered'
output_phrase_table = '/home/chris/projects/angular/editor_components/microservices/lm_autocomplete/test_data/phrase_tables/wmt-phrase-table.en-es.input-filtered'

filtered_lines = []
with codecs.open(full_phrase_table, encoding='utf8') as pt_file:
    for l in pt_file:
        l = l.strip()
        # source_phrase is the first item in each line of the phrase table
        source_phrase = l.split(u' ||| ')[0]
        source_toks = tuple(wordpunct_tokenize(source_phrase))
        
        if source_toks in filter_set:
            filtered_lines.append(l)

with codecs.open(output_phrase_table, 'w', encoding='utf8') as output_file:
    for l in filtered_lines:
        output_file.write(l + '\n')
```
